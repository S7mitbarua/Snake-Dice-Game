```python
#Build a python project about Snake Dice Game.

import random

board = ["", "", "", "", "", "", "", "", "", "X", "", "", "", "", "", "", "", "", "", "O"]
player1 = "X"
player2 = "O"
current_player = player1

while True:
    input("Press Enter to roll the dice...")
    dice_roll = random.randint(1, 6)
    print(f"{current_player} rolled a {dice_roll}!")
    player_index = board.index(current_player)
    new_index = player_index + dice_roll
    if new_index >= len(board):
        print(f"{current_player} has won!")
        break
    if board[new_index] == player1:
        print(f"{player1} has been sent back to the start!")
        board[0] = player1
    elif board[new_index] == player2:
        print(f"{player2} has been sent back to the start!")
        board[10] = player2
    else:
        board[player_index] = ""
        board[new_index] = current_player
        player_index = new_index
    current_player = player1 if current_player == player2 else player2
```
